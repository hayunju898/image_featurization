```python
import SimpleITK as sitk
import numpy as np
import radiomics
import os
import csv
from radiomics import featureextractor
```

# **LOW IMAGES**


```python
# download_directory = "/content/drive/MyDrive/medical_imaging_private/Code/scans/low"

download_directory = "/content/drive/MyDrive/Medical Imaging Meghan Lubner/Spring2024/Deep Learning/Group 1 Project Code/Cutted image/low/"

files = os.listdir(download_directory)

#image filepaths for loop
image_filepaths = []

# for file in files:
#   pathname = '/content/drive/MyDrive/medical_imaging_private/Code/scans/low/' + file
#   image_filepaths.append(pathname)

for file in files:
  pathname = '/content/drive/MyDrive/Medical Imaging Meghan Lubner/Spring2024/Deep Learning/Group 1 Project Code/Cutted image/low/' + file
  image_filepaths.append(pathname)

image_filepaths
```

```python
#feature extractor
params = {
  'binWidth': 25,
  'resampledPixelSpacing': None,
  'interpolator': sitk.sitkBSpline,
  'verbose': True
}

extractor = featureextractor.RadiomicsFeatureExtractor(**params)

feature_vectors = []

for filepath in image_filepaths:
  image = sitk.ReadImage(filepath,sitk.sitkInt8)

  # Otsu's thresholding
  threshold_filter = sitk.OtsuThresholdImageFilter()
  threshold_filter.SetInsideValue(0)
  threshold_filter.SetOutsideValue(1)
  thresholded_image = threshold_filter.Execute(image)

  # binary mask
  binary_mask = sitk.GetArrayFromImage(thresholded_image)

  # convert binary mask to SimpleITK image
  binary_mask_sitk = sitk.GetImageFromArray(binary_mask.astype(np.uint8))

  featureVector = extractor.execute(image, binary_mask_sitk)

  feature_vectors.append(featureVector)

  #creating a mask
  # threshold_value = 100
  # binary_image = sitk.BinaryThreshold(image, lowerThreshold=threshold_value, upperThreshold=255)

  # featureVector = extractor.execute(image, binary_image)
  # feature_vectors.append(featureVector)

#printing out the features
for i, vector in enumerate(feature_vectors):
    print(f"Features extracted from image {i + 1}:")
    for key, value in vector.items():
        print(f"\t{key}: {value}\n")
```

```python
#saving features into a csv
csv_file = 'low_images_features.csv'

with open(csv_file, 'w', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=feature_vectors[0].keys())
    writer.writeheader()

    for vector in feature_vectors:
        writer.writerow(vector)
```
